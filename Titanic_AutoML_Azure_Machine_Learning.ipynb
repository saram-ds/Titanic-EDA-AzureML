## 1. Introduction

Automated Machine Learning (AutoML) is an Artificial Intelligence technology that automates many of the complex and time-consuming tasks involved in developing machine learning models. Instead of manually selecting algorithms, tuning hyperparameters, and comparing model performance, AutoML automatically explores multiple machine learning pipelines to identify the model that best fits the dataset and optimization objective.

Microsoft Azure Machine Learning provides a powerful AutoML service that enables users to build, train, evaluate, and compare machine learning models with minimal coding effort. This significantly accelerates the machine learning development process while allowing both beginners and experienced practitioners to focus on data understanding and business objectives rather than algorithm implementation.

In this project, Azure Machine Learning AutoML was used to develop a regression model for predicting the **Fare** variable from the Titanic dataset. The platform automatically evaluated multiple regression algorithms and selected the best-performing model based on the chosen optimization metric. The project demonstrates how AutoML can simplify model development while maintaining strong predictive performance through automated model selection and evaluation.


## 2. Project Objectives

## 4. Why Regression?

Selecting the appropriate machine learning task is a fundamental step in any predictive analytics project. The choice depends primarily on the nature of the target variable to be predicted.

In this experiment, the target variable is **Fare**, which represents the ticket price paid by each passenger aboard the Titanic. Since Fare is a **continuous numerical variable**, the prediction objective is to estimate a numerical value rather than assign observations to predefined categories. Therefore, the problem is naturally formulated as a **regression task**.

Unlike classification, which predicts discrete labels (e.g., Survived or Not Survived), regression models estimate continuous outcomes. Azure Machine Learning AutoML automatically recognizes this objective and evaluates a collection of regression algorithms to determine the model that best fits the data according to the selected optimization metric.

By defining **Fare** as the target variable and selecting **Regression** as the task type, the AutoML framework was able to compare multiple candidate models and identify the algorithm that achieved the best predictive performance without requiring manual model selection.


## 6. Running the AutoML Experiment

After configuring the AutoML experiment, the training job was submitted to the Azure Machine Learning compute target **titanic-notebook**. Azure ML automatically executed the complete machine learning workflow, including data preprocessing, feature engineering, model training, hyperparameter optimization, and model evaluation.

The experiment started on **July 19, 2026**, and completed successfully after approximately **6 hours and 20 minutes**. During this period, Azure AutoML generated and evaluated numerous candidate pipelines, each combining different preprocessing techniques, regression algorithms, and hyperparameter settings.

Several regression algorithms were explored, including:

- Extreme Random Trees
- Random Forest
- Decision Tree
- ElasticNet
- XGBoost Regressor

In addition to testing different learning algorithms, Azure AutoML also evaluated multiple preprocessing strategies such as **StandardScalerWrapper**, **MaxAbsScaler**, and **SparseNormalizer**. Each pipeline was assessed using the **Normalized Root Mean Squared Error (Normalized RMSE)**, which served as the primary optimization metric for the experiment.

After comparing all candidate models, Azure Machine Learning automatically selected the pipeline that achieved the lowest prediction error as the final model.

## 7. Best Model Evaluation

Once the automated training process was completed, Azure Machine Learning selected the pipeline combining **StandardScalerWrapper** and **Extreme Random Trees** as the best-performing regression model. The selection was based on the optimization of the **Normalized Root Mean Squared Error (Normalized RMSE)**, which served as the primary evaluation metric throughout the AutoML experiment.

The selected model achieved a **Normalized RMSE of 0.06663**, indicating a relatively low prediction error compared with the range of the target variable. Several complementary performance metrics were also computed to provide a more comprehensive assessment of the model's predictive capabilities.

The coefficient of determination (**R² = 0.4835**) indicates that approximately **48% of the variability** in the ticket fare can be explained by the model. While this value suggests a moderate predictive performance, it is reasonable considering the relatively small size of the Titanic dataset and the complexity of the relationships between passenger characteristics and ticket prices.

The **Mean Absolute Error (MAE = 14.37)** and **Median Absolute Error (4.16)** show that the average prediction errors remain within an acceptable range for this dataset. In addition, the **Explained Variance Score (0.4927)** confirms that the model captures nearly half of the observed variability in ticket prices.

Finally, the **Spearman Rank Correlation (0.8336)** demonstrates a strong monotonic relationship between the predicted and actual ticket fares, indicating that the model successfully preserves the relative ordering of observations even when exact numerical predictions are not perfect.

Overall, the evaluation metrics suggest that the selected Extreme Random Trees model provides a reliable predictive performance and represents an appropriate choice for this regression problem.

## 8. Discussion

The AutoML experiment demonstrated the effectiveness of Microsoft Azure Machine Learning in automatically exploring multiple regression pipelines and selecting the most suitable model for the Titanic fare prediction task. Without requiring manual hyperparameter tuning or extensive model engineering, AutoML evaluated numerous combinations of preprocessing techniques and machine learning algorithms before identifying the optimal solution.

An interesting observation is that the highest-ranked models were predominantly based on **tree ensemble algorithms**, particularly **Extreme Random Trees** and **Random Forest**. These algorithms consistently outperformed linear models such as ElasticNet and even gradient boosting approaches evaluated during the experiment. This behavior is expected because ticket fare depends on complex, nonlinear relationships between several passenger characteristics, including passenger class, age, family size, embarkation port, and other categorical attributes. Ensemble tree methods are particularly well suited to capturing these interactions without requiring explicit feature transformations.

The selected pipeline combined **StandardScalerWrapper** with the **Extreme Random Trees** algorithm. Although tree-based methods are generally insensitive to feature scaling, Azure AutoML automatically evaluated different preprocessing strategies as part of its optimization process. This illustrates one of the major advantages of AutoML: preprocessing and model selection are optimized jointly rather than independently.

The evaluation metrics indicate that the model achieved satisfactory predictive performance for this dataset. The relatively low Normalized RMSE, together with a moderate coefficient of determination (R²) and a strong Spearman correlation, suggests that the model produces reliable predictions while preserving the ranking of passenger fares. These results demonstrate that AutoML successfully identified a robust regression pipeline despite the moderate size and inherent complexity of the Titanic dataset.

From a practical perspective, this experiment highlights several advantages of Azure Machine Learning AutoML. It significantly reduces the time required for model development, automatically compares a large number of candidate models, performs hyperparameter optimization, and provides a transparent ranking of all evaluated pipelines. Such capabilities enable data scientists to focus more on data understanding and result interpretation rather than repetitive model tuning.

Nevertheless, some limitations should be acknowledged. The Titanic dataset is relatively small and represents a simplified educational benchmark rather than a real industrial application. Consequently, the obtained performance should not be generalized to more complex prediction problems. Future work could investigate additional feature engineering techniques, alternative validation strategies, and larger real-world datasets to further evaluate the robustness of Azure AutoML.


## 9. Conclusion

This project demonstrated the practical application of Microsoft Azure Machine Learning AutoML for solving a regression problem using the Titanic dataset. Starting from a cleaned dataset prepared during the exploratory data analysis phase, an automated machine learning workflow was configured to identify the most suitable predictive model for estimating passenger ticket fares.

Throughout the experiment, Azure AutoML automatically explored multiple combinations of preprocessing techniques, machine learning algorithms, and hyperparameter configurations. Among all evaluated candidates, the pipeline combining **StandardScalerWrapper** with **Extreme Random Trees** achieved the best overall performance according to the selected optimization metric, Normalized Root Mean Squared Error (Normalized RMSE).

The evaluation results confirmed that the selected model provides satisfactory predictive performance, with low prediction error and a strong monotonic relationship between predicted and actual ticket fares. Beyond model accuracy, this project demonstrated how Azure Machine Learning simplifies the complete machine learning lifecycle by automating model selection, hyperparameter optimization, experiment tracking, and performance comparison.

This experience also highlighted the importance of combining exploratory data analysis with automated machine learning. While AutoML efficiently identifies high-performing models, the quality of the final solution still depends on appropriate data preparation, feature understanding, and careful interpretation of the evaluation metrics.

Overall, this project illustrates how Microsoft Azure Machine Learning can accelerate the development of reliable predictive models while reducing the manual effort traditionally required during the machine learning workflow. The knowledge acquired through this project provides a solid foundation for applying AutoML techniques to larger datasets and more complex real-world predictive analytics problems.
